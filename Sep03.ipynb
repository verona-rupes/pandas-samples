```python
#columns=['TIME', 'TYPE', 'BID', 'ASK', 'BID_SIZE','ASK_SIZE','LAST','LAST_SIZE'];
columns=['TIME', 'TYPE', 'BID', 'ASK'];
df=pd.read_csv('/Users/RAFA/Documents/s3backup/AAPL/AAPL-2016-09-02',sep=',' ,parse_dates=True,dayfirst=True ,names=['DATE', 'TIME', 'TYPE', 'BID', 'ASK', 'BID_SIZE',
'ASK_SIZE','LAST','LAST_SIZE','BIDEXCH','ASKEXCH','LASTEXCH','COND'],usecols=columns)
df.set_index('TIME', inplace = True, drop = False, append = False);
q=df[(df.TYPE == 'QUOTE')]
```

```python
#q['SPREAD']=q['ASK'] - q['BID']
q.loc[:,'SPREAD']=(q['ASK']-q['BID']).round(2)
q.loc[:,'ASK_SPREAD']=(q['ASK'].diff()).round(2)
q.loc[:,'BID_SPREAD']=(q['BID'].diff()).round(2)
q
```
