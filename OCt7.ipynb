```python
# Sort and round 
columns=['TIME','PERM_ID','ORDER_ID','LMT_PRICE','ACTION','SHARES','EXCH','COMM','P/L'];
trades=pd.read_csv('AAPL-Report_Duplicates.log',sep=',',parse_dates=True,dayfirst=True,names=['TIME','PERM_ID','ORDER_ID','LMT_PRICE','ACTION','SHARES','EXCH','COMM','P/L'],usecols=columns)
#dff=df.loc[df['STATUS'].isin(['Submitted', 'Filled','Cancelled'])]
trades_sored=trades.sort_values(['PERM_ID','ORDER_ID'], ascending=[True, True])
trades_sored=trades_sored.round({'COMM': 3, 'P/L': 3})
trades_sored.to_csv('/Users/RAFA/Documents/s3backup/AAPL/logs/AAPL_Trades_Sorted', sep='\t',index=False)
trades_sored.to_html('/Users/RAFA/Documents/s3backup/AAPL/logs/AAPL_Trades_Sorted_html',index=False)
trades_sored
```

```python
# Remove Duplicates
columns=['TIME','PERM_ID','ORDER_ID','LMT_PRICE','ACTION','SHARES','EXCH','COMM','P/L'];
trades=pd.read_csv('AAPL_Trades_Sorted',sep='\t',parse_dates=True,dayfirst=True,names=['TIME','PERM_ID','ORDER_ID','LMT_PRICE','ACTION','SHARES','EXCH','COMM','P/L'],usecols=columns)
#dff=df.loc[df['STATUS'].isin(['Submitted', 'Filled','Cancelled'])]
trades_sorted=trades.sort_values(['PERM_ID','ORDER_ID'], ascending=[True, True])
trades_sorted=trades_sorted.round({'COMM': 6, 'P/L': 6})
trades_sorted=trades_sorted.drop_duplicates()
trades_sorted.to_csv('/Users/RAFA/Documents/s3backup/AAPL/logs/AAPL_Trades_panda', sep='\t',index=False)
trades_sorted.to_html('/Users/RAFA/Documents/s3backup/AAPL/logs/AAPL_Trades_panda_html',index=False)
```
